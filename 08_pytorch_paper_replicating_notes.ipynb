## What is a machine learning research paper?



A machine learning research paper is a scientific paper that details findings of a research group on a specific area.

The contents of a machine learning research paper can vary from paper to paper but they generally follow the structure:

| **Section** | **Contents** |
| ----- | ----- |
| **Abstract** | An overview/summary of the paper's main findings/contributions. |
| **Introduction** | What's the paper's main problem and details of previous methods used to try and solve it. |
| **Method** | How did the researchers go about conducting their research? For example, what model(s), data sources, training setups were used? |
| **Results** | What are the outcomes of the paper? If a new type of model or training setup was used, how did the results of findings compare to previous works? (this is where **experiment tracking** comes in handy) |
| **Conclusion** | What are the limitations of the suggested methods? What are some next steps for the research community? |
| **References** | What resources/other papers did the researchers look at to build their own body of work? |
| **Appendix** | Are there any extra resources/findings to look at that weren't included in any of the above sections? |

## Terminology



There are going to be a fair few acronyms throughout this notebook.

In light of this, here are some definitions:
* **ViT** - Stands for Vision Transformer, a neural network architecture designed for image recognition tasks. Unlike traditional convolutional neural networks (CNNs), ViT applies the Transformer architecture, originally developed for natural language processing (NLP), to image data by dividing images into fixed-size patches and treating them as sequences of tokens. This approach allows ViT to leverage the power of self-attention mechanisms for image classification and other vision tasks.
* **ViT paper** - Refers to the seminal machine learning research paper titled [*An Image is Worth 16x16 Words: Transformers for Image Recognition at Scale*](https://arxiv.org/abs/2010.11929), authored by Dosovitskiy et al. This paper introduced the Vision Transformer (ViT) architecture and demonstrated its effectiveness in image recognition tasks, particularly when trained on large-scale datasets. Anytime *ViT paper* is mentioned, it is referencing this foundational work.

## 3. Replicating the ViT paper: an overview



Before we write any more code, let's discuss what we're doing.

We'd like to replicate the ViT paper for our own problem, FoodVision Mini.

So our **model inputs** are: images of pizza, steak and sushi.

And our ideal **model outputs** are: predicted labels of pizza, steak or sushi.

No different to what we've been doing throughout the previous sections.

The question is: how do we go from our inputs to the desired outputs?

#### ViT Architecture Overview

A Vision Transformer (ViT) is a transformer-based architecture specifically designed for computer vision tasks. Unlike traditional Convolutional Neural Networks (CNNs), which rely on convolutional layers to extract hierarchical features from images, ViTs use a transformer encoder to process image data. Here's a deeper dive into the key concepts:

##### 1. **Patch Extraction and Serialization**
   - **Patch Extraction**: An input image is divided into fixed-size patches (e.g., 16x16 pixels). These patches are analogous to "tokens" in natural language processing (NLP), where a sentence is broken into words or subwords.
   - **Serialization into Vectors**: Each patch is flattened into a 1D vector. For example, a 16x16 patch with 3 color channels (RGB) would be flattened into a vector of size $16 \times 16 \times 3 = 768$. This process is called **serialization** because the 2D patch is converted into a 1D sequence of numbers.
     - **Why Transform into 1D?**: Transforming 2D patches into 1D vectors is necessary because transformers are designed to process sequential data. By flattening the patches, we convert the spatial structure of the image into a format that the transformer can handle. This allows the model to treat image patches similarly to how it treats tokens in NLP, enabling it to learn relationships between patches through self-attention mechanisms.
     - **Why Not Keep It 2D?**: While 2D representations preserve spatial locality, transformers are not inherently designed to process 2D data directly. Flattening the patches simplifies the input format, making it easier to apply linear transformations, matrix multiplications, and other operations that are fundamental to transformer architectures. Additionally, working with 1D vectors allows the model to learn representations that are more flexible and can capture global patterns across the entire image, rather than being constrained by local 2D structures.

##### 2. **Mapping to a Smaller Dimension**
   - After serialization, each patch vector is mapped to a smaller dimension using a **learnable linear projection** (a single matrix multiplication). This step is crucial because:
     - **Reducing Dimensionality**: The dimensionality of the patch embeddings is reduced to make the data more manageable for the transformer encoder. High-dimensional data (e.g., 768 dimensions) can be computationally expensive to process, especially when dealing with large images or datasets. By projecting the patch vectors into a lower-dimensional space (e.g., 512 dimensions), the model reduces the computational load and memory requirements, making training and inference more efficient.
     - **Learning Meaningful Representations**: Mapping to a lower-dimensional space allows the model to learn a more compact and meaningful representation of the patches. In high-dimensional spaces, data points are often sparse, and relationships between them can be harder to capture. By reducing the dimensionality, the model can focus on the most important features of the patches, effectively compressing the information while preserving the essential patterns needed for the task. This is particularly important in vision tasks, where the goal is to extract useful spatial and contextual information from the image patches.
     - **Generalization**: Lower-dimensional representations can also improve the model's ability to generalize. High-dimensional data is prone to overfitting, where the model memorizes the training data instead of learning generalizable patterns. By reducing the dimensionality, the model is forced to focus on the most salient features, which can lead to better performance on unseen data.
   - For example, if the patch vector is 768-dimensional, it might be projected into a 512-dimensional space using a weight matrix of size $768 \times 512$. This weight matrix is learned during training, allowing the model to adaptively determine the best way to compress the patch information while retaining the most relevant features for the task at hand.

##### 3. **Why Transformers Use Tokens**
   - Transformers were originally designed for NLP tasks, where text is inherently sequential. In NLP, a sentence is tokenized into words or subwords, and each token is embedded into a vector.
     - **What is a Token?**: A token is a basic unit of data that represents a meaningful chunk of information. In NLP, a token typically corresponds to a word or a subword (e.g., "cat" or "unhappiness" might be split into "un", "happi", and "ness"). In computer vision, a token can represent an image patch, which is a small region of the image.
     - **Embedding into a Vector**: Each token (word or subword) is mapped to a fixed-size vector in a continuous vector space. For example, the word "cat" might be represented as a 512-dimensional vector like [0.23, -0.45, 0.67, ..., 0.12]. This vector is learned during training and encodes the meaning of the token in a way that the transformer can process.
       - **Encoding Meaning**: The embedding process transforms discrete tokens into continuous numerical representations that capture their semantic meaning. For instance, words with similar meanings (e.g., "cat" and "dog") will have vectors that are closer together in the vector space, while unrelated words (e.g., "cat" and "car") will have vectors farther apart. This allows the transformer to understand relationships between tokens based on their vector representations.
   - In ViTs, **image patches are treated as tokens**. This allows the transformer to process images in a similar way to text, leveraging its ability to model long-range dependencies and relationships between patches. The transformer encoder processes these patch embeddings to capture spatial relationships and patterns in the image.

##### 4. **Comparison to CNNs**
   - **Inductive Bias**: CNNs have a strong inductive bias for local spatial patterns due to their convolutional layers. ViTs, on the other hand, have less inductive bias and rely on the transformer's ability to learn global relationships from the data.
   - **Data Efficiency**: CNNs are generally more data-efficient because of their built-in assumptions about locality and translation invariance. ViTs require larger datasets to achieve similar performance but can scale to much larger models (e.g., 22B parameters).
   - **Capacity**: ViTs have higher capacity due to their transformer architecture, which allows them to model complex relationships between patches.

##### 5. **Applications and Variants**
   - ViTs have been successfully applied to tasks like image classification, object detection, and segmentation. They are also used in domains like weather prediction and autonomous driving.
   - Many hybrid architectures combine ViTs and CNNs to leverage the strengths of both approaches. For example, some models use CNNs for initial feature extraction and ViTs for higher-level reasoning.

##### Summary
- **Serialization**: Flattening image patches into 1D vectors.
- **Mapping**: Reducing patch vector dimensions using a linear projection.
- **Tokens**: Patches are treated as tokens to leverage the transformer's ability to model relationships.
- **Transformers vs CNNs**: ViTs trade off data efficiency for higher capacity and flexibility.

This enhanced explanation clarifies the key steps and motivations behind the ViT architecture.


#### **CNN vs. Transformer: Key Differences**


##### **Convolutional Neural Network (CNN)**  
- **Definition**:  
  A neural network architecture designed for processing grid-like data (e.g., images) using **convolutional layers** that apply learnable filters to local regions of the input.  

- **Key Features**:  
  - **Inductive Bias**: Built-in assumptions about spatial locality and translation invariance (e.g., edges/textures matter regardless of position).  
  - **Hierarchical Feature Learning**: Lower layers detect edges/textures → higher layers detect complex patterns (e.g., shapes, objects).  
  - **Weight Sharing**: The same filter is applied across all spatial positions, reducing parameters.  

- **Use Cases**:  
  - Image classification (e.g., ResNet, VGG).  
  - Object detection (e.g., YOLO, Faster R-CNN).  
  - Medical imaging (e.g., tumor detection).  

- **Examples**:  
  - **ResNet-50**: Classifies 1000 ImageNet categories.  
  - **U-Net**: Segmenting tumors in MRI scans.  

- **Advantages**:  
  - Efficient for spatially local patterns.  
  - Requires less training data due to inductive biases.  
  - Computationally lighter for high-resolution images.  

- **Limitations**:  
  - Struggles with long-range dependencies (e.g., global context).  
  - Fixed receptive fields limit flexibility.  

---

##### **Transformer**  
- **Definition**:  
  A neural network architecture using **self-attention mechanisms** to model relationships between all elements in a sequence, regardless of distance.  

- **Key Features**:  
  - **Self-Attention**: Computes pairwise interactions between all tokens (e.g., patches in an image).  
  - **No Built-In Spatial Bias**: Relies on positional embeddings to encode spatial/temporal relationships.  
  - **Global Context**: Captures long-range dependencies in one layer.  

- **Use Cases**:  
  - Natural Language Processing (e.g., BERT, GPT).  
  - Vision tasks (e.g., ViT for image classification).  
  - Multimodal tasks (e.g., CLIP for text-image alignment).  

- **Examples**:  
  - **ViT (Vision Transformer)**: Classifies images by splitting them into patches.  
  - **BERT**: Understands context in sentences.  

- **Advantages**:  
  - Excels at modeling global interactions.  
  - Flexible architecture (handles variable-length inputs).  
  - Parallelizable computation (no sequential dependency).  

- **Limitations**:  
  - Computationally expensive for long sequences (quadratic attention complexity).  
  - Requires large datasets to learn spatial relationships from scratch.  

---

##### **Comparison Table**  
| **Aspect**               | **CNN**                                  | **Transformer**                          |  
|--------------------------|------------------------------------------|------------------------------------------|  
| **Core Operation**       | Convolution (local)                      | Self-attention (global)                  |  
| **Inductive Bias**       | Strong (spatial locality)                | Weak (learns from data)                  |  
| **Data Efficiency**      | High (works well with small datasets)    | Low (needs large datasets)               |  
| **Computational Cost**   | Scales with image size ($O(n)$)          | Scales with sequence length² ($O(n²)$)   |  
| **Long-Range Context**   | Limited (requires deep stacks)           | Built-in (captured in one layer)         |  
| **Example Models**       | ResNet, MobileNet                        | ViT, BERT, DETR                          |  

---

##### **When to Use Which?**  
- **Choose CNN** if:  
  - Working with small-to-medium datasets.  
  - Processing high-resolution images (e.g., 4K video).  
  - Tasks benefit from local patterns (e.g., edge detection).  

- **Choose Transformer** if:  
  - Data is abundant (e.g., web-scale datasets).  
  - Global context is critical (e.g., scene understanding).  
  - Multimodal or sequential data (e.g., video + text).  

- **Hybrid Models** (e.g., ConvNeXt, Swin Transformer): Combine CNN’s efficiency with Transformer’s global reasoning.  


### 3.1 Inputs and outputs, layers and blocks



ViT is a deep learning neural network architecture.

And any neural network architecture is generally comprised of **layers**.

And a collection of layers is often referred to as a **block**.

And stacking many blocks together is what gives us the whole architecture.

A **layer** takes an input (say an image tensor), performs some kind of function on it (for example what's in the layer's `forward()` method) and then returns an output.

So if a **single layer** takes an input and gives an output, then a collection of layers or a **block** also takes an input and gives an output.

Let's make this concrete:
* **Layer** - takes an input, performs a function on it, returns an output.
* **Block** - a collection of layers, takes an input, performs a series of functions on it, returns an output.
* **Architecture (or model)** - a collection of blocks, takes an input, performs a series of functions on it, returns an output.

This ideology is what we're going to be using to replicate the ViT paper.

We're going to take it layer by layer, block by block, function by function putting the pieces of the puzzle together like Lego to get our desired overall architecture.

The reason we do this is because looking at a whole research paper can be intimidating.

So for a better understanding, we'll break it down, starting with the inputs and outputs of single layer and working up to the inputs and outputs of the whole model.

The ViT architecture is comprised of several stages:

* **Patch + Position Embedding (inputs)** -
  - **Patches**: In the context of the Vision Transformer (ViT), an image is divided into smaller, fixed-size patches. These patches are essentially smaller sub-images that are extracted from the original image. For example, if you have a 224x224 image and you divide it into 16x16 patches, you will end up with 196 patches (224/16 = 14, so 14x14 = 196 patches). Each patch is then flattened into a vector, which becomes the input to the transformer.
  - **Position Embedding**: Since transformers do not inherently understand the spatial structure of the image (unlike CNNs), position embeddings are added to the patches to provide information about the original position of each patch in the image. This helps the model understand the spatial relationships between different patches. Position embeddings are typically learned during training and are added to the patch embeddings before they are fed into the transformer.

* **Linear projection of flattened patches (Embedded Patches)** -
  - **Embedded Patches**: After the image is divided into patches and flattened, each patch is represented as a vector of pixel values. However, these raw pixel values may not be the most effective representation for the model to learn from. To address this, a **linear projection** is applied to each patch. This means that each patch vector is multiplied by a learned weight matrix (a linear transformation) to map it into a higher-dimensional space. The purpose of this projection is to transform the raw pixel values into a more meaningful, learnable representation. This new representation, often referred to as an "embedding," captures more abstract features of the patch, which can be useful for tasks like image classification. The embedding process allows the model to learn relationships between patches that are not immediately apparent from the raw pixel values.
    - **Linear Transformation**: A linear transformation is a mathematical operation that maps a vector from one space to another using a matrix multiplication. In this case, the transformation is applied to the flattened patch vectors. The transformation matrix is learned during training, meaning the model adjusts the weights of the matrix to optimize the representation of the patches.
    - **Higher-Dimensional Space**: The raw pixel values of a patch are typically in a lower-dimensional space (e.g., 16x16 = 256 dimensions for a 16x16 patch). By applying a linear transformation, the patch is mapped to a higher-dimensional space (e.g., 768 dimensions). This higher-dimensional space allows the model to capture more complex and abstract features of the patch, which are not directly observable in the raw pixel values. This is essentially increasing the number of parameters (weights) in the model that allow the model to capture more nuanced and absract features that might not be apparent in the raw pixel values.
    - **Meaningful Representation**: The linear transformation helps to encode the patch in a way that is more useful for the model. For example, it might capture patterns like edges, textures, or other visual features that are important for distinguishing between different classes of images. This transformation is crucial because it allows the model to learn from the data in a way that is more aligned with the task at hand (e.g., image classification).


* **Norm** -
  - **Layer Normalization (Norm)**: Layer normalization is a technique used to stabilize and speed up the training of deep neural networks. It normalizes the activations of a layer across the features (as opposed to batch normalization, which normalizes across the batch dimension). This means that for each sample in the batch, the activations are normalized to have a mean of zero and a standard deviation of one. Layer normalization is particularly useful in transformers because it helps to reduce the internal covariate shift, which can lead to faster convergence and more stable training. In PyTorch, you can use [`torch.nn.LayerNorm()`](https://pytorch.org/docs/stable/generated/torch.nn.LayerNorm.html) to apply layer normalization.
    - **What does it mean to normalize?**: Normalization in machine learning refers to scaling or transforming the data so that it has a consistent range or distribution. This is often done to make the optimization process (e.g., gradient descent) more efficient and stable. In the context of neural networks, normalization typically involves adjusting the activations (outputs of a layer) so that they have a mean of 0 and a standard deviation of 1.
    - **Why a mean of 0 and a standard deviation of 1?**: Normalizing activations to have a mean of 0 and a standard deviation of 1 helps to center the data around zero and scale it to a consistent range. This is beneficial because:
      - **Stable Gradients**: It prevents the gradients from becoming too large or too small, which can cause instability during training (e.g., exploding or vanishing gradients).
      - **Faster Convergence**: By keeping the activations in a consistent range, the optimization algorithm (e.g., gradient descent) can converge more quickly to a good solution.
      - **Better Generalization**: Normalization can also help the model generalize better to unseen data by reducing the impact of outliers or extreme values.
    - **What is internal covariate shift?**: Internal covariate shift refers to the change in the distribution of the activations of a layer during training. As the weights of the network are updated, the distribution of the inputs to each layer can change, making it harder for the network to learn effectively. Normalization helps to mitigate this by keeping the distribution of activations stable, which in turn makes the training process more stable and efficient.
    - **What is convergence?**: Convergence in machine learning refers to the process by which the model's parameters (weights) stabilize and the loss function reaches a minimum (or near-minimum) value. When a model converges, it means that further training is unlikely to significantly improve performance. Faster convergence is desirable because it means the model can learn effectively in fewer training steps.
    - **Why does reducing internal covariate shift lead to more stable training?**: By reducing internal covariate shift, normalization ensures that the distribution of activations remains relatively stable throughout training. This stability allows the optimizer to make more consistent updates to the model's weights, leading to faster and more stable training. Without normalization, the optimizer might struggle to find the right direction for weight updates, leading to slower convergence or even instability in training.


* **Multi-Head Attention (MSA)** -
  - **Self-Attention**: A mechanism where each element in a sequence (e.g., a patch embedding) computes a weighted sum over all other elements. The weights ("attention scores") determine how much focus to place on other positions when encoding the current position. This allows the model to dynamically highlight relevant patches (e.g., "pizza crust" patches might attend to "cheese" patches).
  - **Multi-Head**: Instead of one attention mechanism, multi-head attention runs *multiple independent attention mechanisms ("heads") in parallel*. Each head projects the input into a different **representation subspace** (a lower-dimensional space via learned linear transformations) to focus on different types of relationships (e.g., one head for spatial proximity, another for color/texture). The outputs of all heads are concatenated and linearly projected to form the final output. This parallel processing enables the model to capture diverse patterns.
  - **Why It Works**: By attending to all patches simultaneously, self-attention handles long-range dependencies better than convolutional kernels with limited receptive fields. Multi-head design increases model capacity and interpretability (different heads can learn distinct behaviors).

* **MLP (Multilayer Perceptron)** -
  - **Structure**: A 2-layer neural network with `[Linear → GELU → Dropout → Linear → Dropout]`. The first linear layer expands the dimension (e.g., 768 → 3072 in ViT-Base), adding model capacity, while the second layer compresses back to the original dimension.
  - **Role**: After self-attention (which mixes information *across patches*), the MLP processes features *within each patch embedding*. It introduces non-linearity (via GELU) and allows hierarchical feature learning.
  - **GELU vs ReLU**: Gaussian Error Linear Unit (GELU) is smoother than ReLU, providing better gradient flow: `GELU(x) = x * Φ(x)` where Φ(x) is the Gaussian CDF. This approximates a "soft" ReLU and often performs better in transformers.
  - **Why It Works**: The MLP provides a learnable non-linear transformation that refines the features produced by attention. Expansion/contraction helps learn complex interactions without blowing up dimensions.

* **Transformer Encoder** -
  - **Structure**: A repeated block of `[LayerNorm → MSA → Residual Connection → LayerNorm → MLP → Residual Connection]`. Multiple encoder blocks are stacked (e.g., 12 times in ViT-Base).
  - **Residual Connections**: The input is added to the output of MSA/MLP (`output = input + module(input)`). This mitigates vanishing gradients by creating "shortcut paths" for gradient flow, enabling deeper networks.
  - **Pre-LayerNorm vs Post-LayerNorm**: ViT uses **Pre-LayerNorm** (LayerNorm *before* MSA/MLP), which stabilizes training compared to the original Transformer's Post-LayerNorm. This is why LayerNorm appears at the start of each block.
  - **Why It Works**: Stacking encoder blocks creates a deep hierarchy where lower layers capture local patterns (edges, textures) and higher layers capture global semantics (e.g., "pizza" vs "sushi"). Residual connections ensure stable gradient propagation through many layers.

* **MLP Head** -
  - **Structure**: Typically a single linear layer (or a small MLP) that maps the final `[class]` token embedding to class logits. In ViT, this is `[LayerNorm → Linear]` applied only to the `[class]` token (other patch tokens are discarded).
  - **[class] Token**: A learnable embedding prepended to the patch embeddings. During processing, it aggregates global image information via self-attention (other patches "report" to it). The final `[class]` token serves as the image representation.
  - **Why It Works**: The `[class]` token acts as a master query that distills information from all patches. A simple linear classifier works because the transformer encoder already produces a highly discriminative embedding.

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-figure-1-inputs-and-outputs-food-mini.png" width=900 alt="figure 1 from the original vision transformer paper adapted to work with food images, an image of pizza goes in and gets classified as 'pizza'"/>

*The complete processing flow: Each encoder block alternates between mixing information globally (MSA) and transforming features locally (MLP), progressively refining the [class] token's representation until the MLP head makes the final prediction.*

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-mapping-the-four-equations-to-figure-1.png" width=1000 alt="mapping the vision transformer paper figure 1 to the four equations listed in the paper"/>

There's a lot happening in the image above but following the coloured lines and arrows reveals the main concepts of the ViT architecture.

How about we break down each equation further (it will be our goal to recreate these with code)?

In all equations (except equation 4), "$\mathbf{z}$" is the raw output of a particular layer:

1. $\mathbf{z}_{0}$ is "z zero" (this is the output of the initial patch embedding layer).
2. $\mathbf{z}_{\ell}^{\prime}$ is "z of a particular layer *prime*" (or an intermediary value of z).
3. $\mathbf{z}_{\ell}$ is "z of a particular layer".

And $\mathbf{y}$ is the overall output of the architecture.

#### 3.2.3 Equation 1 overview

$$
\begin{aligned}
\mathbf{z}_{0} &=\left[\mathbf{x}_{\text {class }} ; \mathbf{x}_{p}^{1} \mathbf{E} ; \mathbf{x}_{p}^{2} \mathbf{E} ; \cdots ; \mathbf{x}_{p}^{N} \mathbf{E}\right]+\mathbf{E}_{\text {pos }}, & & \mathbf{E} \in \mathbb{R}^{\left(P^{2} \cdot C\right) \times D}, \mathbf{E}_{\text {pos }} \in \mathbb{R}^{(N+1) \times D}
\end{aligned}
$$

This equation deals with the class token, patch embedding and position embedding ($\mathbf{E}$ is for embedding) of the input image.

In vector form, the embedding might look something like:

```python
x_input = [class_token, image_patch_1, image_patch_2, image_patch_3...] + [class_token_position, image_patch_1_position, image_patch_2_position, image_patch_3_position...]
```

Where each of the elements in the vector is learnable (their `requires_grad=True`).

#### 3.2.4 Equation 2 overview

$$
\begin{aligned}
\mathbf{z}_{\ell}^{\prime} &=\operatorname{MSA}\left(\operatorname{LN}\left(\mathbf{z}_{\ell-1}\right)\right)+\mathbf{z}_{\ell-1}, & & \ell=1 \ldots L
\end{aligned}
$$

This says that for every layer from $1$ through to $L$ (the total number of layers), there's a Multi-Head Attention layer (MSA) wrapping a LayerNorm layer (LN).

The addition on the end is the equivalent of adding the input to the output and forming a [skip/residual connection](https://paperswithcode.com/method/residual-connection).

We'll call this layer the "MSA block".

In pseudocode, this might look like:

```python
x_output_MSA_block = MSA_layer(LN_layer(x_input)) + x_input
```

Notice the skip connection on the end (adding the input of the layers to the output of the layers).

#### 3.2.5 Equation 3 overview

$$
\begin{aligned}
\mathbf{z}_{\ell} &=\operatorname{MLP}\left(\operatorname{LN}\left(\mathbf{z}_{\ell}^{\prime}\right)\right)+\mathbf{z}_{\ell}^{\prime}, & & \ell=1 \ldots L \\
\end{aligned}
$$

This says that for every layer from $1$ through to $L$ (the total number of layers), there's also a Multilayer Perceptron layer (MLP) wrapping a LayerNorm layer (LN).

The addition on the end is showing the presence of a skip/residual connection.

We'll call this layer the "MLP block".

In pseudocode, this might look like:

```python
x_output_MLP_block = MLP_layer(LN_layer(x_output_MSA_block)) + x_output_MSA_block
```

Notice the skip connection on the end (adding the input of the layers to the output of the layers).

#### 3.2.6 Equation 4 overview

$$
\begin{aligned}
\mathbf{y} &=\operatorname{LN}\left(\mathbf{z}_{L}^{0}\right) & &
\end{aligned}
$$

This says for the last layer $L$, the output $y$ is the 0 index token of $z$ wrapped in a LayerNorm layer (LN).

Or in our case, the 0 index of `x_output_MLP_block`:

```python
y = Linear_layer(LN_layer(x_output_MLP_block[0]))
```

Of course there are some simplifications above but we'll take care of those when we start to write PyTorch code for each section.

## 4. Equation 1: Split data into patches and creating the class, position and patch embedding



I remember one of my machine learning engineer friends used to say "it's all about the embedding."

As in, if you can represent your data in a good, learnable way (as **embeddings are learnable representations**), chances are, a learning algorithm will be able to perform well on them.

With that being said, let's start by creating the class, position and patch embeddings for the ViT architecture.

We'll start with the **patch embedding**.

This means we'll be turning our input images in a sequence of patches and then embedding those patches.

Recall that an **embedding** is a learnable representation of some form and is often a vector.

The term learnable is important because this means the numerical representation of an input image (that the model sees) can be improved over time.

We'll begin by following the opening paragraph of section 3.1 of the ViT paper (bold mine):

> The standard Transformer receives as input a 1D sequence of token embeddings. To handle 2D images, we reshape the image $\mathbf{x} \in \mathbb{R}^{H \times W \times C}$ into a sequence of flattened 2D patches $\mathbf{x}_{p} \in \mathbb{R}^{N \times\left(P^{2} \cdot C\right)}$, where $(H, W)$ is the resolution of the original image, $C$ is the number of channels, $(P, P)$ is the resolution of each image patch, and $N=H W / P^{2}$ is the resulting number of patches, which also serves as the effective input sequence length for the Transformer. The Transformer uses constant latent vector size $D$ through all of its layers, so we flatten the patches and map to $D$ dimensions with a trainable linear projection (Eq. 1). We refer to the output of this projection as the **patch embeddings**.

And size we're dealing with image shapes, let's keep in mind the line from Table 3 of the ViT paper:

> Training resolution is **224**.

Let's break down the text above.

* $D$ is the size of the **patch embeddings**, different values for $D$ for various sized ViT models can be found in Table 1.
* The image starts as 2D with size ${H \times W \times C}$.
    * $(H, W)$ is the resolution of the original image (height, width).
    * $C$ is the number of channels.
* The image gets converted to a sequence of flattened 2D patches with size ${N \times\left(P^{2} \cdot C\right)}$.
    * $(P, P)$ is the resolution of each image patch (**patch size**).
    * $N=H W / P^{2}$ is the resulting number of patches, which also serves as the input sequence length for the Transformer.

### 4.1 Calculating patch embedding input and output shapes by hand



How about we start by calculating these input and output shape values by hand?

To do so, let's create some variables to mimic each of the terms (such as $H$, $W$ etc) above.

We'll use a **patch size ($P$) of 16 since it's the best performing version of ViT-Base uses** (see column "ViT-B/16" of Table 5 in the ViT paper for more).

````python
# Create example values
height = 224 # H ("The training resolution is 224.")
width = 224 # W
color_channels = 3 # C
patch_size = 16 # P

# Calculate N (number of patches)
# This determines how many patches of size (P, P) can fit into an image of size (H, W)
# N = HW/P^2
number_of_patches = int((height * width) / patch_size**2)

print(f"Number of patches (N) with image height (H={height}), width (W={width}) and patch size (P={patch_size}): {number_of_patches}")
````

We've got the number of patches, how about we create the image output size as well?

Better yet, let's replicate the input and output shapes of the patch embedding layer.

Recall:

* **Input:** The image starts as 2D with size ${H \times W \times C}$.
* **Output:** The image gets converted to a sequence of flattened 2D patches with size ${N \times\left(P^{2} \cdot C\right)}$

````python
# Input shape (this is the size of a single image)
# H x W x C
embedding_layer_input_shape = (height, width, color_channels)

# Output shape (this is the size of the embedding)
# N x (P^2 * C)
embedding_layer_output_shape = (number_of_patches, patch_size ** 2 * color_channels)

print(f"Input shape (single 2D image): {embedding_layer_input_shape}")
print(f"Output shape (single 2D image flattened into patches): {embedding_layer_output_shape}")
````

### 4.3 Creating image patches with `torch.nn.Conv2d()`



We've seen what an image looks like when it gets turned into patches, now let's start moving towards replicating the patch embedding layers with PyTorch.

To visualize our single image we wrote code to loop through the different height and width dimensions of a single image and plot individual patches.

This operation is very similar to the convolutional operation we saw in [03. PyTorch Computer Vision section 7.1: Stepping through `nn.Conv2d()`](https://www.learnpytorch.io/03_pytorch_computer_vision/#71-stepping-through-nnconv2d).

In fact, the authors of the ViT paper mention in section 3.1 that the patch embedding is achievable with a convolutional neural network (CNN):  

> **Hybrid Architecture.** As an alternative to raw image patches, the input sequence can be formed from feature maps of a CNN (LeCun et al., 1989). In this hybrid model, the patch embedding projection $\mathbf{E}$ (Eq. 1) is applied to patches extracted from a **CNN feature map**. As a special case, the patches can have spatial size $1 \times 1$, which means that the **input sequence is obtained by simply flattening the spatial dimensions of the feature map and projecting to the Transformer dimension**. The classification input embedding and position embeddings are added as described above.

The "**feature map**" they're referring to are the weights/activations produced by a convolutional layer passing over a given image.

<img src="https://github.com/mrdbourke/pytorch-deep-learning/raw/main/images/08-vit-paper-patch-embedding-animation.gif" width=900 alt="example of creating a patch embedding by passing a convolutional layer over a single image"/>

*By setting the `kernel_size` and `stride` parameters of a [`torch.nn.Conv2d()`](https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html) layer equal to the `patch_size`, we can effectively get a layer that splits our image into patches and creates a learnable embedding (referred to as a "Linear Projection" in the ViT paper) of each patch.*

Remember our ideal input and output shapes for the patch embedding layer?

* **Input:** The image starts as 2D with size ${H \times W \times C}$.
* **Output:** The image gets converted to a 1D sequence of flattened 2D patches with size ${N \times\left(P^{2} \cdot C\right)}$.

Or for an image size of 224 and patch size of 16:

* **Input (2D image):** (224, 224, 3) -> (height, width, color channels)
* **Output (flattened 2D patches):** (196, 768) -> (number of patches, embedding dimension)

We can recreate these with:
* [`torch.nn.Conv2d()`](https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html) for turning our image into patches of CNN feature maps.
* [`torch.nn.Flatten()`](https://pytorch.org/docs/stable/generated/torch.nn.Flatten.html) for flattening the spatial dimensions of the feature map.

Let's start with the `torch.nn.Conv2d()` layer.

We can replicate the creation of patches by setting the `kernel_size` and `stride` equal to `patch_size`.

This means each convolutional kernel will be of size `(patch_size x patch_size)` or if `patch_size=16`, `(16 x 16)` (the equivalent of one whole patch).

And each step or `stride` of the convolutional kernel will be `patch_size` pixels long or `16` pixels long (equivalent of stepping to the next patch).

We'll set `in_channels=3` for the number of color channels in our image and we'll set `out_channels=768`, the same as the $D$ value in Table 1 for ViT-Base (this is the embedding dimension, each image will be embedded into a learnable vector of size 768).

````python
from torch import nn

# Set the patch size
patch_size=16

# Create the Conv2d layer with hyperparameters from the ViT paper
conv2d = nn.Conv2d(in_channels=3, # number of color channels
                   out_channels=768, # from Table 1: Hidden size D, this is the embedding size
                   kernel_size=patch_size, # could also use (patch_size, patch_size)
                   stride=patch_size,  # non-overlapping patches
                   padding=0)
````

Now we've got a convolutional layer, let's see what happens when we pass a single image through it.

````python
# View single image
plt.imshow(image.permute(1, 2, 0)) # adjust for matplotlib
plt.title(class_names[label])
plt.axis(False);
````

conv2d expects a 4D tensor of shape `(batch_size, in_channels, height, width)` so we need to add a batch dimension to our image tensor using `unsqueeze(0)`.

````python
# Pass the image through the convolutional layer
image_out_of_conv = conv2d(image.unsqueeze(0)) # add a single batch dimension (height, width, color_channels) -> (batch, height, width, color_channels)
print(image_out_of_conv.shape)
````

Passing our image through the convolutional layer turns it into a series of 768 (this is the embedding size or $D$) feature/activation maps.

So its output shape can be read as:
    
```python
torch.Size([1, 768, 14, 14]) -> [batch_size, embedding_dim, feature_map_height, feature_map_width]
```

Let's visualize five random feature maps and see what they look like.

````python
# Plot random 5 convolutional feature maps
import random
random_indexes = random.sample(range(0, 758), k=5) # pick 5 numbers between 0 and the embedding size
print(f"Showing random convolutional feature maps from indexes: {random_indexes}")

# Create plot
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(12, 12))

# Plot random image feature maps
for i, idx in enumerate(random_indexes):
    image_conv_feature_map = image_out_of_conv[:, idx, :, :] # index on the output tensor of the convolutional layer
    axs[i].imshow(image_conv_feature_map.squeeze().detach().numpy())
    axs[i].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[]);
````

#### Recap of the patch embedding layer

The embedding dimension in the context of the Vision Transformer (ViT) refers to the size of the vector that represents each patch after it has been processed by the `PatchEmbedding` layer. Let's break this down step by step:

1. **Patch Creation**:
   - Given an image of size $H \times W$ and a patch size of $P \times P$, the image is divided into $N$ patches, where $N = \frac{H \times W}{P^2}$.
   - For example, if you have an image of size $224 \times 224$ and a patch size of $16 \times 16$, the number of patches $N$ would be $\frac{224 \times 224}{16 \times 16} = 196$.

2. **Patch Embedding**:
   - Each patch is then flattened into a 1D vector. Since each patch is $16 \times 16$ and has 3 color channels (RGB), the flattened vector for each patch will have a size of $16 \times 16 \times 3 = 768$.
   - This 768-dimensional vector is then passed through a linear projection (a fully connected layer) to map it to an embedding dimension $D$. In the case of ViT-Base, $D = 768$.

3. **Embedding Dimension**:
   - The embedding dimension $D$ is the size of the vector that represents each patch after the linear projection. In ViT-Base, this is 768.
   - So, each patch is represented by a 768-dimensional vector, and the entire sequence of patches (including the class token) will have a shape of $(N + 1) \times 768$, where $N$ is the number of patches and $+1$ accounts for the class token.

4. **Position Embedding**:
   - Position embeddings are added to the patch embeddings to retain positional information. These position embeddings also have a dimension of $D = 768$.

To summarize:
- The embedding dimension $D$ is the size of the vector that represents each patch after it has been processed by the `PatchEmbedding` layer.
- In ViT-Base, $D = 768$, meaning each patch is represented by a 768-dimensional vector.
- The total number of patches $N$ is calculated based on the image size and patch size, and the sequence of patch embeddings (including the class token) will have a shape of $(N + 1) \times 768$.



### 4.5 Turning the ViT patch embedding layer into a PyTorch module



Time to put everything we've done for creating the patch embedding into a single PyTorch layer.

We can do so by subclassing `nn.Module` and creating a small PyTorch "model" to do all of the steps above.

Specifically we'll:
1. Create a class called `PatchEmbedding` which subclasses `nn.Module` (so it can be used a PyTorch layer).
2. Initialize the class with the parameters `in_channels=3`, `patch_size=16` (for ViT-Base) and `embedding_dim=768` (this is $D$ for ViT-Base from Table 1).
3. Create a layer to turn an image into patches using `nn.Conv2d()` (just like in 4.3 above).
4. Create a layer to flatten the patch feature maps into a single dimension (just like in 4.4 above).
5. Define a `forward()` method to take an input and pass it through the layers created in 3 and 4.
6. Make sure the output shape reflects the required output shape of the ViT architecture (${N \times\left(P^{2} \cdot C\right)}$).

Let's do it!

````python
# 1. Create a class which subclasses nn.Module
class PatchEmbedding(nn.Module):
    """Turns a 2D input image into a 1D sequence learnable embedding vector.

    Args:
        in_channels (int): Number of color channels for the input images. Defaults to 3.
        patch_size (int): Size of patches to convert input image into. Defaults to 16.
        embedding_dim (int): Size of embedding to turn image into. Defaults to 768.
    """
    # 2. Initialize the class with appropriate variables
    def __init__(self,
                 in_channels:int=3, # 3 b/c RGB image
                 patch_size:int=16, # 16 b/c ViT uses 16x16 patches
                 embedding_dim:int=768):    # 768 b/c 16*16*3 = 768
        super().__init__()

        # 3. Create a layer to turn an image into patches
        self.patcher = nn.Conv2d(in_channels=in_channels,
                                 out_channels=embedding_dim,    # learnable vector D
                                 kernel_size=patch_size,
                                 stride=patch_size,
                                 padding=0)

        # 4. Create a layer to flatten the patch feature maps into a single dimension
        self.flatten = nn.Flatten(start_dim=2, # only flatten the feature map dimensions into a single vector
                                  end_dim=3)

    # 5. Define the forward method
    def forward(self, x):   # self is the object itself, x is the input image
        # Create assertion to check that inputs are the correct shape
        image_resolution = x.shape[-1]  # -1 takes last dim of tensor, which is width
        assert image_resolution % patch_size == 0, f"Input image size must be divisible by patch size, image shape: {image_resolution}, patch size: {patch_size}"

        # Perform the forward pass
        x_patched = self.patcher(x)
        x_flattened = self.flatten(x_patched)
        # 6. Make sure the output shape has the right order
        return x_flattened.permute(0, 2, 1) # adjust so the embedding is on the final dimension [batch_size, P^2•C, N] -> [batch_size, N, P^2•C]
````



#### Explanation of the Code and Its Relation to the ViT Paper

This code defines a `PatchEmbedding` class that is responsible for converting a 2D input image into a sequence of 1D patch embeddings, which is a crucial step in the Vision Transformer (ViT) architecture. Let's break it down step by step and relate it to the ViT paper.

##### 1. **Class Definition and Initialization**
   - The `PatchEmbedding` class inherits from `nn.Module`, which is the base class for all neural network modules in PyTorch.
   - The `__init__` method initializes the class with the following parameters:
     - `in_channels`: The number of color channels in the input image (default is 3 for RGB images).
     - `patch_size`: The size of the patches to which the image will be divided (default is 16, as used in the ViT paper).
     - `embedding_dim`: The dimension of the embedding vector for each patch (default is 768, as used in the ViT-Base model).

##### 2. **Creating the Patch Embedding Layer**
   - The `self.patcher` layer is a 2D convolutional layer (`nn.Conv2d`) that is used to divide the input image into patches and linearly embed each patch into a vector of size `embedding_dim`.
     - `in_channels`: Number of input channels (3 for RGB images).
     - `out_channels`: Number of output channels (equal to `embedding_dim`).
     - `kernel_size`: Size of the convolutional kernel (equal to `patch_size`). This ensures that each patch is processed independently, as the kernel will cover the entire patch without overlapping.
     - `stride`: Stride of the convolution (equal to `patch_size`). This ensures that the convolution operation moves exactly one patch at a time, preventing overlap between patches and ensuring that the image is divided into non-overlapping patches.
     - `padding`: Padding size (0 to ensure no padding is added, which would otherwise alter the patch boundaries).

   This convolutional layer effectively splits the image into patches of size `patch_size x patch_size` and projects each patch into a `embedding_dim`-dimensional space. By setting both the `kernel_size` and `stride` equal to the `patch_size`, we ensure that:
   - Each patch is processed independently, as the kernel covers the entire patch.
   - The convolution operation moves exactly one patch at a time, ensuring that the image is divided into non-overlapping patches.
   - The output of this layer will have a shape of `[batch_size, embedding_dim, num_patches_height, num_patches_width]`, where `num_patches_height` and `num_patches_width` are determined by the input image resolution divided by the `patch_size`.


##### 3. **Flattening the Patches**
   - The `self.flatten` layer is a `nn.Flatten` layer that flattens the patch feature maps into a single dimension. This is necessary because the Transformer expects a sequence of 1D vectors as input.
     - `start_dim=2`: Start flattening from the second dimension (the patch height dimension).
     - `end_dim=3`: End flattening at the third dimension (the patch width dimension).

   After flattening, the output shape will be `[batch_size, num_patches, embedding_dim]`, where `num_patches` is the number of patches obtained from the image.

##### 4. **Forward Method**
   - The `forward` method defines the forward pass of the `PatchEmbedding` layer.
   - **Assertion Check**: The method first checks if the input image's resolution is divisible by the `patch_size`. This ensures that the image can be evenly divided into patches without any remainder.
   - **Patch Embedding**: The input image is passed through the `self.patcher` layer to obtain the patch embeddings.
   - **Flattening**: The patch embeddings are then flattened using the `self.flatten` layer.
   - **Permutation**: Finally, the output is permuted to ensure that the embedding dimension is on the last axis, resulting in the shape `[batch_size, num_patches, embedding_dim]`.

3. **Position Embedding**:
   - $\mathbf{E}_{\text{pos}}$: This is a learnable positional embedding that is added to the patch embeddings (and the class token) to provide positional information to the Transformer. Since Transformers are permutation-invariant, position embeddings are necessary to encode the spatial structure of the image.
   - The position embedding is also **not yet implemented** in the code. It needs to be added as a learnable parameter and combined with the patch embeddings and class token.

```python
patch_embedding = [image_patch_1, image_patch_2, image_patch_3...]

```

Notice the concatenation (`torch.cat()`) happens on `dim=1` (the `number_of_patches` dimension).

Let's create a learnable embedding for the class token.

To do so, we'll get the batch size and embedding dimension shape and then we'll create a `torch.ones()` tensor in the shape `[batch_size, 1, embedding_dimension]`.

And we'll make the tensor learnable by passing it to `nn.Parameter()` with `requires_grad=True`.

````python
# Get the batch size and embedding dimension
# remember: patch_embedded_image.shape = [batch_size, num_patches, embedding_dimension]
batch_size = patch_embedded_image.shape[0]
embedding_dimension = patch_embedded_image.shape[2]

# # Create the class token embedding as a learnable parameter that shares the same size as the embedding dimension (D)
# class_token = nn.Parameter(torch.ones(batch_size, 1, embedding_dimension),  # [batch_size, number_of_tokens, embedding_dimension]
#                            requires_grad=True)  # gradient is required for backpropagation

# using randn instead of ones
# Create the class token embedding as a learnable parameter that shares the same size as the embedding dimension (D)
class_token = nn.Parameter(torch.randn(batch_size, 1, embedding_dimension),  # [batch_size, number_of_tokens, embedding_dimension]
                           requires_grad=True)  # gradient is required for backpropagation

# Show the first 10 examples of the class token
print(class_token[:, :, :10])

# Print the class_token shape
print(f"Class token shape: {class_token.shape} -> [batch_size, number_of_tokens, embedding_dimension]")
````

Now we've got the class token embedding, let's prepend it to our sequence of image patches, `patch_embedded_image`.

We can do so using [`torch.cat()`](https://pytorch.org/docs/stable/generated/torch.cat.html) and set `dim=1` (so `class_token`'s `number_of_tokens` dimension is prepended to `patch_embedded_image`'s `number_of_patches` dimension).

````python
# Add the class token embedding to the front of the patch embedding
patch_embedded_image_with_class_embedding = torch.cat((class_token, patch_embedded_image),
                                                      dim=1) # concat on first dimension

# Print the sequence of patch embeddings with the prepended class token embedding
print(patch_embedded_image_with_class_embedding)
print(f"The shape of the patch embedded image before the class token: {patch_embedded_image.shape} -> [batch_size, number_of_patches, embedding_dimension]")
print(f"Sequence of patch embeddings with class token prepended shape: {patch_embedded_image_with_class_embedding.shape} -> [batch_size, number_of_patches, embedding_dimension]")
````

`dim=1` means that the concatenation will happen along the second dimension (the `number_of_patches` dimension).

### 5.1 The LayerNorm (LN) Layer

**What is Layer Normalization?**
- **Definition**: Layer Normalization (`torch.nn.LayerNorm`) is a technique that normalizes the input data over the last dimension. This is commonly used in neural networks to stabilize and speed up training.
- **Purpose**: It helps improve training efficiency and model generalization by ensuring that the data samples have similar distributions (similar mean and standard deviation).
- **Formal Definition**: You can find the formal definition of `torch.nn.LayerNorm()` in the [PyTorch documentation](https://pytorch.org/docs/stable/generated/torch.nn.LayerNorm.html).

**Why is Layer Normalization Important?**
- **Improves Training Time**: By normalizing the data, the model can converge faster during training.
- **Enhances Generalization**: Normalized data helps the model perform better on unseen data by reducing internal covariate shift.
- **Analogy**: Think of normalization as making all the stairs in a staircase the same height and length. Just as it’s easier to walk up stairs of uniform size, neural networks can optimize more efficiently over data with consistent distributions.

**How Does Layer Normalization Work?**
- **Normalized Shape**: The key parameter in `torch.nn.LayerNorm()` is `normalized_shape`, which specifies the dimension size to normalize over. For example, in ViT-Base, this would be `768` (the embedding dimension).
- **Normalization Process**:
  - **Step 1**: Compute the mean ($\mu$) and standard deviation ($\sigma$) of the input data over the specified dimension.
    - Formula:
      $$
      \mu = \frac{1}{H} \sum_{i=1}^{H} x_i, \quad \sigma = \sqrt{\frac{1}{H} \sum_{i=1}^{H} (x_i - \mu)^2}
      $$
      where $H$ is the size of the dimension being normalized, and $x_i$ are the elements of the input tensor.
  - **Step 2**: Normalize the data by subtracting the mean and dividing by the standard deviation.
    - Formula:
      $$
      \hat{x}_i = \frac{x_i - \mu}{\sigma}
      $$
  - **Step 3**: Apply a learned scale ($\gamma$) and shift ($\beta$) to the normalized data to allow the model to adapt the normalization if needed.
    - Formula:
      $$
      y_i = \gamma \hat{x}_i + \beta
      $$
      where $\gamma$ and $\beta$ are learnable parameters.

**Visual Explanation**
- **Before Normalization**: Imagine a dataset where each sample has a different mean and standard deviation. This variability makes it harder for the model to learn effectively.
- **After Normalization**: All samples are transformed to have a similar mean and standard deviation, making it easier for the model to optimize and generalize.

**Example Scenario**
- **Before LayerNorm**:
  - Data samples have varying distributions (e.g., some samples have high values, others have low values).
  - Example Tensor:
    ```python file.py
    # Input tensor before normalization
    input_tensor = torch.tensor([[1.0, 2.0, 3.0],
                                 [10.0, 20.0, 30.0],
                                 [100.0, 200.0, 300.0]])
    ```
    - The values in this tensor vary significantly across rows, making it harder for the model to learn.
  - The model struggles to adjust its weights effectively, leading to slower convergence.
- **After LayerNorm**:
  - All samples are scaled to have a mean of 0 and a standard deviation of 1.
  - Example Tensor:
    ```python file.py
    # Output tensor after normalization
    output_tensor = torch.tensor([[-1.2247,  0.0,  1.2247],
                                  [-1.2247,  0.0,  1.2247],
                                  [-1.2247,  0.0,  1.2247]])
    ```
    - The values are now standardized, with each row having a mean of 0 and a standard deviation of 1.
  - The model can now learn more efficiently, as the data is in a consistent format.

**Key Takeaways**
- **Normalization**: LayerNorm ensures that the input data is standardized, making it easier for the model to learn.
- **Efficiency**: It reduces the time required for the model to converge during training.
- **Generalization**: By stabilizing the input distribution, the model performs better on unseen data.

**Further Reading**
- [Layer Normalization Paper](https://arxiv.org/abs/1607.06450)
- [PyTorch LayerNorm Documentation](https://pytorch.org/docs/stable/generated/torch.nn.LayerNorm.html)

### Multi-Head Self-Attention (MSA) Layer

**What is Multi-Head Self-Attention (MSA)?**
- **Definition**:
  - MSA is a mechanism that allows a model to focus on different parts of an input sequence simultaneously. It is a core component of the Transformer architecture, introduced in the [*Attention is All You Need*](https://arxiv.org/abs/1706.03762) paper.
  - **Purpose**: MSA enables the model to capture relationships between different elements of the input sequence (e.g., words in a sentence or patches in an image) by computing attention scores across multiple "heads."
  - **Formal Definition**: The MSA layer is implemented in PyTorch using `torch.nn.MultiheadAttention()`. You can find the formal documentation [here](https://pytorch.org/docs/stable/generated/torch.nn.MultiheadAttention.html).

**Why is MSA Important?**
- **Captures Complex Relationships**: MSA allows the model to learn intricate dependencies between different parts of the input, which is crucial for tasks like language translation, image classification, and more.
- **Parallel Processing**: By using multiple attention heads, the model can process different parts of the input sequence in parallel, improving efficiency.
- **Flexibility**: MSA can be applied to various types of data, including text, images, and even audio.

**How Does MSA Work?**
- **Input**: The input to the MSA layer is typically a sequence of embeddings (e.g., word embeddings for text or patch embeddings for images).
- **Query, Key, Value (QKV)**: The input is split into three components:
  - **Query (Q)**: Represents the current element being processed.
  - **Key (K)**: Represents the elements to which the query will attend.
  - **Value (V)**: Represents the information that will be aggregated based on the attention scores.
- **Attention Mechanism**: The attention scores are computed as:
  $$
  \text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V
  $$
  where $d_k$ is the dimension of the key vectors.
- **Multi-Head Attention**: Instead of computing a single attention score, MSA computes multiple attention scores in parallel using different sets of Q, K, and V matrices. These are then concatenated and linearly transformed:
  $$
  \text{MultiHead}(Q, K, V) = \text{Concat}(\text{head}_1, \text{head}_2, \dots, \text{head}_h)W^O
  $$
  where each head is computed as:
  $$
  \text{head}_i = \text{Attention}(QW_i^Q, KW_i^K, VW_i^V)
  $$
  and $W_i^Q$, $W_i^K$, $W_i^V$, and $W^O$ are learned projection matrices.

**Why Call It "Multi-Head"?**
- **Multiple Attention Heads**: The term "multi-head" refers to the use of multiple attention mechanisms (or "heads") that operate in parallel. Each head learns to focus on different aspects of the input sequence.
  - **Example**: In a sentence, one head might focus on syntactic relationships (e.g., subject-verb agreement), while another head might focus on semantic relationships (e.g., word meanings).
- **Increased Capacity**: By using multiple heads, the model can capture a richer set of relationships, leading to better performance on complex tasks.

**What is a Head?**
- **Definition**: A "head" in MSA refers to a single attention mechanism that computes attention scores for a specific aspect of the input sequence.
- **Function**: Each head independently computes its own set of Q, K, and V matrices, allowing it to focus on different parts of the input.
- **Example**: In an image, one head might focus on edges, while another head might focus on textures.

**Visual Explanation**
- **Before MSA**: Imagine an image divided into patches. Without MSA, the model might struggle to understand how different patches relate to each other.
- **After MSA**: The model can now compute attention scores between patches, allowing it to understand relationships like "this patch is part of the same object as that patch."

**Example Scenarios**

  - With MSA, the model can compute attention scores to understand that "dog" relates strongly to "jumped" and "fence."

### 5.3 Replicating Equation 2 with PyTorch layers



Let's put everything we've discussed about the LayerNorm (LN) and Multi-Head Attention (MSA) layers in equation 2 into practice.

To do so, we'll:  

1. Create a class called `MultiheadSelfAttentionBlock` that inherits from `torch.nn.Module`.
2. Initialize the class with hyperparameters from Table 1 of the ViT paper for the ViT-Base model.
3. Create a layer normalization (LN) layer with `torch.nn.LayerNorm()` with the `normalized_shape` parameter the same as our embedding dimension ($D$ from Table 1).
4. Create a multi-head attention (MSA) layer with the appropriate `embed_dim`, `num_heads`, `dropout` and `batch_first` parameters.
5. Create a `forward()` method for our class passing the in the inputs through the LN layer and MSA layer.

````python
# 1. Create a class that inherits from nn.Module
class MultiheadSelfAttentionBlock(nn.Module):
    """Creates a multi-head self-attention block ("MSA block" for short).
    """
    # 2. Initialize the class with hyperparameters from Table 1
    def __init__(self,
                 embedding_dim:int=768, # Hidden size D from Table 1 for ViT-Base
                 num_heads:int=12, # Heads from Table 1 for ViT-Base
                 attn_dropout:float=0): # doesn't look like the paper uses any dropout in MSABlocks
        super().__init__()

        # 3. Create the Norm layer (LN)
        self.layer_norm = nn.LayerNorm(normalized_shape=embedding_dim)

        # 4. Create the Multi-Head Attention (MSA) layer
        self.multihead_attn = nn.MultiheadAttention(embed_dim=embedding_dim,
                                                    num_heads=num_heads,
                                                    dropout=attn_dropout,
                                                    batch_first=True) # does our batch dimension come first?

    # 5. Create a forward() method to pass the data through the layers
    def forward(self, x):
        x = self.layer_norm(x)
        attn_output, _ = self.multihead_attn(query=x, # query embeddings
                                             key=x, # key embeddings
                                             value=x, # value embeddings
                                             need_weights=False) # do we need the weights or just the layer outputs?
        return attn_output
````

## 6. Equation 3: Multilayer Perceptron (MLP)

**What is MLP?**
- **Multilayer Perceptron (MLP)**: A class of feedforward artificial neural network (ANN) that consists of at least three layers of nodes: an input layer, one or more hidden layers, and an output layer. Each layer is fully connected to the next one, and each node (except for the input nodes) is a neuron that uses a nonlinear activation function.
- **Role in Transformer Encoder**: In the Vision Transformer (ViT) architecture, the MLP block is used to process the output of the Multi-Head Self-Attention (MSA) block, applying non-linear transformations to capture complex patterns in the data.

**Why is MLP Important?**
- **Non-Linearity**: MLPs introduce non-linearities through activation functions, allowing the model to learn complex mappings between inputs and outputs.
- **Feature Transformation**: The MLP block transforms the features extracted by the MSA block, enabling the model to capture higher-level abstractions.
- **Residual Connection**: The addition of the residual connection (skip connection) helps in mitigating the vanishing gradient problem, allowing for deeper networks and more stable training.

**How Does MLP Work?**
- **Equation 3**:
  $$
  \mathbf{z}_{\ell} = \operatorname{MLP}\left(\operatorname{LN}\left(\mathbf{z}_{\ell}^{\prime}\right)\right) + \mathbf{z}_{\ell}^{\prime}, \quad \ell=1 \ldots L
  $$
  - **Layer Normalization (LN)**: The input to the MLP block is first normalized using Layer Normalization.
  - **MLP Transformation**: The normalized input is then passed through a series of fully connected layers with non-linear activation functions (e.g., GELU).
  - **Residual Connection**: The output of the MLP block is added to the original input (before normalization), forming a residual connection.

**What is a Perceptron?**
- **Perceptron**: The simplest form of a neural network, consisting of a single layer of neurons. Each neuron computes a weighted sum of its inputs, applies an activation function, and produces an output.
- **Multilayer Perceptron (MLP)**: Extends the perceptron by adding multiple layers of neurons, allowing the network to learn more complex functions.

**Structure of the MLP Block in ViT**
- The term [MLP](https://en.wikipedia.org/wiki/Multilayer_perceptron) is quite broad as it can refer to almost any combination of *multiple* layers (hence the "multi" in multilayer perceptron).
- But it generally follows the pattern of:
  - `linear layer -> non-linear layer -> linear layer -> non-linear layer`
- In the case of the ViT paper, the MLP structure is defined in section 3.1:
  - The MLP contains two layers with a GELU non-linearity.
  - Where "two layers" refers to linear layers ([`torch.nn.Linear()`](https://pytorch.org/docs/stable/generated/torch.nn.Linear.html) in PyTorch) and "GELU non-linearity" is the GELU (Gaussian Error Linear Units) non-linear activation function ([`torch.nn.GELU()`](https://pytorch.org/docs/stable/generated/torch.nn.GELU.html) in PyTorch).
- Another sneaky detail about the MLP block doesn't appear until Appendix B.1 (Training):
  - Table 3 summarizes our training setups for our different models. ...Dropout, when used, is applied **after every dense layer except for the the qkv-projections and directly after adding positional- to patch embeddings.**
  - This means that every linear layer (or dense layer) in the MLP block has a dropout layer ([`torch.nn.Dropout()`](https://pytorch.org/docs/stable/generated/torch.nn.Dropout.html) in PyTorch).
  - The value of which can be found in Table 3 of the ViT paper (for ViT-Base, `dropout=0.1`).
- Knowing this, the structure of our MLP block will be:
  - `layer norm -> linear layer -> non-linear layer -> dropout -> linear layer -> dropout`
- With hyperparameter values for the linear layers available from Table 1 (MLP size is the number of hidden units between the linear layers and hidden size $D$ is the output size of the MLP block).

### 6.2 Replicating Equation 3 with PyTorch layers



Let's put everything we've discussed about the LayerNorm (LN) and MLP (MSA) layers in equation 3 into practice.

To do so, we'll:  

1. Create a class called `MLPBlock` that inherits from `torch.nn.Module`.
2. Initialize the class with hyperparameters from Table 1 and Table 3 of the ViT paper for the ViT-Base model.
3. Create a layer normalization (LN) layer with `torch.nn.LayerNorm()` with the `normalized_shape` parameter the same as our embedding dimension ($D$ from Table 1).
4. Create a sequential series of MLP layers(s) using `torch.nn.Linear()`, `torch.nn.Dropout()` and `torch.nn.GELU()` with appropriate hyperparameter values from Table 1 and Table 3.
5. Create a `forward()` method for our class passing the in the inputs through the LN layer and MLP layer(s).

````python
# 1. Create a class that inherits from nn.Module
class MLPBlock(nn.Module):
    """Creates a layer normalized multilayer perceptron block ("MLP block" for short)."""
    # 2. Initialize the class with hyperparameters from Table 1 and Table 3
    def __init__(self,
                 embedding_dim:int=768, # Hidden Size D from Table 1 for ViT-Base
                 mlp_size:int=3072, # MLP size from Table 1 for ViT-Base
                 dropout:float=0.1): # Dropout from Table 3 for ViT-Base
        super().__init__()

        # 3. Create the Norm layer (LN)
        self.layer_norm = nn.LayerNorm(normalized_shape=embedding_dim)

        # 4. Create the Multilayer perceptron (MLP) layer(s)
        self.mlp = nn.Sequential(
            nn.Linear(in_features=embedding_dim,
                      out_features=mlp_size),
            nn.GELU(), # "The MLP contains two layers with a GELU non-linearity (section 3.1)."
            nn.Dropout(p=dropout),
            nn.Linear(in_features=mlp_size, # needs to take same in_features as out_features of layer above
                      out_features=embedding_dim), # take back to embedding_dim
            nn.Dropout(p=dropout) # "Dropout, when used, is applied after every dense layer.."
        )

    # 5. Create a forward() method to pass the data through the layers
    def forward(self, x):
        x = self.layer_norm(x)
        x = self.mlp(x)
        return x
````

## 7. The Transformer Encoder

- **What is a Transformer Encoder?**
  - In deep learning, an **encoder** (or autoencoder) refers to a stack of layers that transforms an input into a numerical representation.
  - In the context of the **Vision Transformer (ViT)**, the **Transformer Encoder** is a key component that processes the input (in this case, a patched image embedding) using alternating layers of:
    - **Multihead Self-Attention (MSA) blocks**: These help the model focus on different parts of the input simultaneously.
    - **MLP (Multi-Layer Perceptron) blocks**: These are fully connected layers that further process the data.
  - The Transformer Encoder is inspired by the architecture introduced in the paper [*Attention is All You Need*](https://arxiv.org/abs/1706.03762), which was originally designed for natural language processing (NLP) but adapted for computer vision in the ViT paper.

- **How does the Transformer Encoder work in ViT?**
  - The ViT paper describes the Transformer Encoder as a series of alternating layers:
    - **Multihead Self-Attention (MSA) blocks**: These compute attention scores between different patches of the image.
    - **MLP blocks**: These apply non-linear transformations to the output of the MSA blocks.
  - **Layer Normalization (LN)** is applied before every block to stabilize training.
  - **Residual connections** are added after every block to help with gradient flow and improve training stability.

- **What are Residual Connections?**
  - **Residual connections** (also called skip connections) were introduced in the paper [*Deep Residual Learning for Image Recognition*](https://arxiv.org/abs/1512.03385v1).
  - They work by adding the input of a layer directly to its output, effectively creating a "shortcut" around the layer.
  - In the ViT architecture:
    - The input to the MSA block is added to the output of the MSA block before passing it to the MLP block.
    - Similarly, the input to the MLP block is added to the output of the MLP block before passing it to the next Transformer Encoder block.
  - In pseudocode:
    ```
    x_input -> MSA_block -> [MSA_block_output + x_input] -> MLP_block -> [MLP_block_output + MSA_block_output + x_input] -> ...
    ```

- **Why do Residual Connections Matter?**
  - **Prevent Vanishing Gradients**: Residual connections help prevent gradients from becoming too small during backpropagation, which is especially important in deep networks.
  - **Enable Deeper Networks**: By allowing gradients to flow more easily, residual connections enable the training of much deeper networks, which can learn more complex representations.
  - **Improve Training Stability**: They stabilize training by ensuring that even if a layer’s output is not useful, the input can still pass through unchanged.

- **Why is this Important for ViT?**
  - The ViT architecture relies on stacking multiple Transformer Encoder blocks to process image data effectively.
  - Without residual connections, training such deep networks would be much harder due to vanishing gradients and instability.
  - Residual connections are a key reason why architectures like ResNet (which introduced residual connections) and ViT can achieve state-of-the-art performance in computer vision tasks.

### 7.1 Creating a Transformer Encoder by combining our custom made layers



Enough talk, let's see this in action and make a ViT Transformer Encoder with PyTorch by combining our previously created layers.

To do so, we'll:  

1. Create a class called `TransformerEncoderBlock` that inherits from `torch.nn.Module`.
2. Initialize the class with hyperparameters from Table 1 and Table 3 of the ViT paper for the ViT-Base model.
3. Instantiate a MSA block for equation 2 using our `MultiheadSelfAttentionBlock` from section 5.2 with the appropriate parameters.
4. Instantiate a MLP block for equation 3 using our `MLPBlock` from section 6.2 with the appropriate parameters.
5. Create a `forward()` method for our `TransformerEncoderBlock` class.
6. Create a residual connection for the MSA block (for equation 2).
7. Create a residual connection for the MLP block (for equation 3).

````python
# 1. Create a class that inherits from nn.Module
class TransformerEncoderBlock(nn.Module):
    """Creates a Transformer Encoder block."""
    # 2. Initialize the class with hyperparameters from Table 1 and Table 3
    def __init__(self,
                 embedding_dim:int=768, # Hidden size D from Table 1 for ViT-Base
                 num_heads:int=12, # Heads from Table 1 for ViT-Base
                 mlp_size:int=3072, # MLP size from Table 1 for ViT-Base
                 mlp_dropout:float=0.1, # Amount of dropout for dense layers from Table 3 for ViT-Base
                 attn_dropout:float=0): # Amount of dropout for attention layers
        super().__init__()

        # 3. Create MSA block (equation 2)
        self.msa_block = MultiheadSelfAttentionBlock(embedding_dim=embedding_dim,
                                                     num_heads=num_heads,
                                                     attn_dropout=attn_dropout)

        # 4. Create MLP block (equation 3)
        self.mlp_block =  MLPBlock(embedding_dim=embedding_dim,
                                   mlp_size=mlp_size,
                                   dropout=mlp_dropout)

    # 5. Create a forward() method
    def forward(self, x):

        # 6. Create residual connection for MSA block (add the input to the output)
        x =  self.msa_block(x) + x

        # 7. Create residual connection for MLP block (add the input to the output)
        x = self.mlp_block(x) + x

        return x
````

### 7.2 Creating a Transformer Encoder with PyTorch's Transformer layers



So far we've built the components of and the Transformer Encoder layer itself ourselves.

But because of their rise in popularity and effectiveness, PyTorch now has in-built [Transformer layers as part of `torch.nn`](https://pytorch.org/docs/stable/nn.html#transformer-layers).

For example, we can recreate the `TransformerEncoderBlock` we just created using [`torch.nn.TransformerEncoderLayer()`](https://pytorch.org/docs/stable/generated/torch.nn.TransformerEncoderLayer.html#torch.nn.TransformerEncoderLayer) and setting the same hyperparameters as above.

You might be thinking, "if we could create the Transformer Encoder so quickly with PyTorch layers, why did we bother reproducing equation 2 and 3?"

The answer is: practice.

Now we've replicated a series of equations and layers from a paper, if you need to change the layers and try something different you can.

But there are benefits of using the PyTorch pre-built layers, such as:
* **Less prone to errors** - Generally, if a layer makes it into the PyTorch standard library, it's been tested and tried to work.
* **Potentially better performance** - As of July 2022 and PyTorch 1.12, the PyTorch implemented version of `torch.nn.TransformerEncoderLayer()` can see [a speedup of more than 2x on many common workloads](https://pytorch.org/blog/a-better-transformer-for-fast-transformer-encoder-inference/).

Finally, since the ViT architecture uses several Transformer Layers stacked on top of each for the full architecture (Table 1 shows 12 Layers in the case of ViT-Base), you can do this with [`torch.nn.TransformerEncoder(encoder_layer, num_layers)`](https://pytorch.org/docs/stable/generated/torch.nn.TransformerEncoder.html#torch.nn.TransformerEncoder) where:
* `encoder_layer` - The target Transformer Encoder layer created with `torch.nn.TransformerEncoderLayer()`.
* `num_layers` - The number of Transformer Encoder layers to stack together.

## 8. Putting it all together to create ViT



Alright, alright, alright, we've come a long way!

But now it's time to do the exciting thing of putting together all of the pieces of the puzzle.

We're going to combine all of the blocks we've created to replicate the full ViT architecture.

From the patch and positional embedding to the Transformer Encoder(s) to the MLP Head.

But wait, we haven't created equation 4 yet...

$$

\end{aligned}
$$

Don't worry, we can put equation 4 into our overall ViT architecture class.

All we need is a `torch.nn.LayerNorm()` layer and a `torch.nn.Linear()` layer to convert the 0th index ($\mathbf{z}_{L}^{0}$) of the Transformer Encoder logit outputs to the target number of classes we have.

To create the full architecture, we'll also need to stack a number of our `TransformerEncoderBlock`s on top of each other, we can do this by passing a list of them to `torch.nn.Sequential()` (this will make a sequential range of `TransformerEncoderBlock`s).

We'll focus on the ViT-Base hyperparameters from Table 1 but our code should be adaptable to other ViT variants.

Creating ViT will be our biggest code block yet but we can do it!

Finally, to bring our own implementation of ViT to life, let's:

1. Create a class called `ViT` that inherits from `torch.nn.Module`.
2. Initialize the class with hyperparameters from Table 1 and Table 3 of the ViT paper for the ViT-Base model.
3. Make sure the image size is divisible by the patch size (the image should be split into even patches).
4. Calculate the number of patches using the formula $N=H W / P^{2}$, where $H$ is the image height, $W$ is the image width and $P$ is the patch size.
5. Create a learnable class embedding token (equation 1) as done above in section 4.6.
6. Create a learnable position embedding vector (equation 1) as done above in section 4.7.
7. Setup the embedding dropout layer as discussed in Appendix B.1 of the ViT paper.
8. Create the patch embedding layer using the `PatchEmbedding` class as above in section 4.5.
9. Create a series of Transformer Encoder blocks by passing a list of `TransformerEncoderBlock`s created in section 7.1 to `torch.nn.Sequential()` (equations 2 & 3).
10. Create the MLP head (also called classifier head or equation 4) by passing a `torch.nn.LayerNorm()` (LN) layer and a `torch.nn.Linear(out_features=num_classes)` layer (where `num_classes` is the target number of classes) linear layer to `torch.nn.Sequential()`.
11. Create a `forward()` method that accepts an input.
12. Get the batch size of the input (the first dimension of the shape).
13. Create the patching embedding using the layer created in step 8 (equation 1).
14. Create the class token embedding using the layer created in step 5 and expand it across the number of batches found in step 11 using [`torch.Tensor.expand()`](https://pytorch.org/docs/stable/generated/torch.Tensor.expand.html) (equation 1).
15. Concatenate the class token embedding created in step 13 to the first dimension of the patch embedding created in step 12 using [`torch.cat()`](https://pytorch.org/docs/stable/generated/torch.cat.html) (equation 1).
16. Add the position embedding created in step 6 to the patch and class token embedding created in step 14 (equation 1).
17. Pass the patch and position embedding through the dropout layer created in step 7.
18. Pass the patch and position embedding from step 16 through the stack of Transformer Encoder layers created in step 9 (equations 2 & 3).
19. Pass index 0 of the output of the stack of Transformer Encoder layers from step 17 through the classifier head created in step 10 (equation 4).
20. Dance and shout woohoo!!! We just built a Vision Transformer!

You ready?

Let's go.

````python
# 1. Create a ViT class that inherits from nn.Module
class ViT(nn.Module):
    """Creates a Vision Transformer architecture with ViT-Base hyperparameters by default."""
    # 2. Initialize the class with hyperparameters from Table 1 and Table 3
    def __init__(self,
                 img_size:int=224, # Training resolution from Table 3 in ViT paper
                 in_channels:int=3, # Number of channels in input image
                 patch_size:int=16, # Patch size
                 num_transformer_layers:int=12, # Layers from Table 1 for ViT-Base
                 embedding_dim:int=768, # Hidden size D from Table 1 for ViT-Base
                 mlp_size:int=3072, # MLP size from Table 1 for ViT-Base
                 num_heads:int=12, # Heads from Table 1 for ViT-Base
                 attn_dropout:float=0, # Dropout for attention projection
                 mlp_dropout:float=0.1, # Dropout for dense/MLP layers
                 embedding_dropout:float=0.1, # Dropout for patch and position embeddings
                 num_classes:int=1000): # Default for ImageNet but can customize this
        super().__init__() # don't forget the super().__init__()!

        # 3. Make the image size is divisible by the patch size
        assert img_size % patch_size == 0, f"Image size must be divisible by patch size, image size: {img_size}, patch size: {patch_size}."

        # 4. Calculate number of patches (height * width/patch^2)
        self.num_patches = (img_size * img_size) // patch_size**2

        # 5. Create learnable class embedding (needs to go at front of sequence of patch embeddings)
        self.class_embedding = nn.Parameter(data=torch.randn(1, 1, embedding_dim),  # [batch_size, number_of_tokens, embedding_dimension]
                                            requires_grad=True)

        # 6. Create learnable position embedding
        self.position_embedding = nn.Parameter(data=torch.randn(1,
                                                                self.num_patches+1, # N = HW/P^2 (# of patches)
                                                                embedding_dim),  # Epos = (N+1) x D (D is size of patch embedding)
                                               requires_grad=True)

        # 7. Create embedding dropout value
        self.embedding_dropout = nn.Dropout(p=embedding_dropout)

        # 8. Create patch embedding layer
        self.patch_embedding = PatchEmbedding(in_channels=in_channels,
                                              patch_size=patch_size,
                                              embedding_dim=embedding_dim)

        # 9. Create Transformer Encoder blocks (we can stack Transformer Encoder blocks using nn.Sequential())
        # Note: The "*" means "all"
        # Note: TransformerEncoderBlock() is a custom class that has MSA and MLP blocks inside it
        self.transformer_encoder = nn.Sequential(*[TransformerEncoderBlock(embedding_dim=embedding_dim,
                                                                           num_heads=num_heads,
                                                                           mlp_size=mlp_size,
                                                                           mlp_dropout=mlp_dropout) for _ in range(num_transformer_layers)])

        # 10. Create classifier head
        self.classifier = nn.Sequential(
            nn.LayerNorm(normalized_shape=embedding_dim),
            nn.Linear(in_features=embedding_dim,
                      out_features=num_classes)
        )

    # 11. Create a forward() method
    def forward(self, x):

        # 12. Get batch size
        batch_size = x.shape[0]

        # 13. Create class token embedding and expand it to match the batch size (equation 1)
        class_token = self.class_embedding.expand(batch_size, -1, -1) # "-1" means to infer the dimension (try this line on its own)

        # 14. Create patch embedding (equation 1)
        x = self.patch_embedding(x)


        # 15. Concat class embedding and patch embedding (equation 1)
        x = torch.cat((class_token, x), dim=1)

        # 16. Add position embedding to patch embedding (equation 1)
        x = self.position_embedding + x

        # 17. Run embedding dropout (Appendix B.1)
        x = self.embedding_dropout(x)

        # 18. Pass patch, position and class embedding through transformer encoder layers (equations 2 & 3)
        x = self.transformer_encoder(x)

        # 19. Put 0 index logit through classifier (equation 4)
        x = self.classifier(x[:, 0]) # run on each sample in a batch at 0 index

        return x
````

##### 9. Create Transformer Encoder blocks (we can stack Transformer Encoder blocks using nn.Sequential())
```python
self.transformer_encoder = nn.Sequential(*[TransformerEncoderBlock(embedding_dim=embedding_dim,
                                                                   num_heads=num_heads,
                                                                   mlp_size=mlp_size,
                                                                   mlp_dropout=mlp_dropout) for _ in range(num_transformer_layers)])
```
- This line creates a sequence of Transformer encoder blocks. The `nn.Sequential` function is used to stack multiple `TransformerEncoderBlock` layers.
  - **Why use `*`?** The `*` operator unpacks the list comprehension into individual arguments for `nn.Sequential`. Omitting it would result in passing a single list object instead of separate layers, causing a TypeError. Proper unpacking ensures each block is treated independently, maintaining modularity and clarity in architecture definition [[3]].
    - **Example Code:**
      ```python
      layer1 = TransformerEncoderBlock(...)
      layer2 = TransformerEncoderBlock(...)
      layer3 = TransformerEncoderBlock(...)

      # Incorrect way (causes TypeError)
      layers = [layer1, layer2, layer3]
      nn.Sequential(layers)

      # Correct way (unpacks layers properly)
      nn.Sequential(layer1, layer2, layer3)
      nn.Sequential(*layers)
      ```
  - **Structure of TransformerEncoderBlock:** Typically comprises Multi-Head Self-Attention followed by Feed-Forward Networks, interspersed with normalization and residual connections. These components collaboratively enable the model to capture intricate dependencies and hierarchies within the input data.

- **Why Stack Multiple Transformer Encoder Layers?**
  - **Depth and Complexity:** Similar to stacking multiple convolutional layers in CNNs, stacking multiple Transformer Encoder layers allows the model to learn increasingly complex patterns and representations. Each layer refines the features extracted from the previous one, enabling deeper understanding and abstraction of the input data.
  - **Hierarchical Feature Learning:** Early layers might capture basic features or relationships, while deeper layers focus on more nuanced and high-level interactions. This hierarchical learning enhances the model's ability to generalize across various tasks and datasets.
  - **Residual Connections and Normalization:** Incorporating residual connections and normalization layers within each `TransformerEncoderBlock` facilitates stable training and gradient flow through deep networks. These mechanisms mitigate issues like vanishing gradients, allowing effective training even with numerous stacked layers.

- **Where Are Multiple Transformer Encoders Used in the Code?**
  - **List Comprehension to Generate Layers:**
    ```python
    transformer_encoder_blocks = [
        TransformerEncoderBlock(
            embedding_dim=embedding_dim,
            num_heads=num_heads,
            mlp_size=mlp_size,
            mlp_dropout=mlp_dropout
        ) for _ in range(num_transformer_layers)
    ]
    self.transformer_encoder = nn.Sequential(*transformer_encoder_blocks)
    ```
    - In this snippet, `num_transformer_layers` determines how many `TransformerEncoderBlock` instances are created. Each block is appended to the `transformer_encoder_blocks` list.
  
  - **Using `nn.Sequential` to Stack Layers:**
    - After generating the list of `TransformerEncoderBlock` layers, `nn.Sequential` stacks them sequentially. The `*` operator unpacks the list so that `nn.Sequential` receives each block as an individual argument rather than a single list object.
    - **Illustrative Example:**
      ```python
      # Define three separate Transformer Encoder Blocks
      block1 = TransformerEncoderBlock(embedding_dim=768, num_heads=12, mlp_size=3072, mlp_dropout=0.1)
      block2 = TransformerEncoderBlock(embedding_dim=768, num_heads=12, mlp_size=3072, mlp_dropout=0.1)
      block3 = TransformerEncoderBlock(embedding_dim=768, num_heads=12, mlp_size=3072, mlp_dropout=0.1)

      # Incorrect usage (results in TypeError)
      blocks_list = [block1, block2, block3]
      sequential_blocks_incorrect = nn.Sequential(blocks_list)

      # Correct usage (using * to unpack the list)
      sequential_blocks_correct = nn.Sequential(block1, block2, block3)
      sequential_blocks_correct_unpacked = nn.Sequential(*blocks_list)
      ```

By stacking multiple Transformer Encoder layers, the ViT architecture leverages the strengths of deep learning to build sophisticated feature representations, ensuring robust performance across diverse applications and datasets.

##### 13. Create class token embedding and expand it to match the batch size (equation 1)
```python
class_token = self.class_embedding.expand(batch_size, -1, -1) # "-1" means to infer the dimension
```
- This line creates the class token embedding and expands it to match the batch size.
  - **Why `-1`?** The `-1` tells PyTorch to infer the size of that dimension from the original tensor. For example, if `self.class_embedding` has shape `[1, 1, embedding_dim]`, the `-1` ensures the last two dimensions remain unchanged while expanding the first dimension to `batch_size`.
  
  - **What does `expand()` do?**
    - The `expand()` function replicates the class token along the batch dimension without allocating extra memory. For instance, if `batch_size = 3` and `embedding_dim = 4`, the original shape `[1, 1, 4]` becomes `[3, 1, 4]`. Each sample in the batch gets its own copy of the class token.

  - **Why is this needed?**
    - In a batched setting, every input image requires its own class token for processing. Expanding ensures each sample in the batch has an independent class token while maintaining efficient memory usage.

By using `expand()`, the code ensures scalability and efficiency, making it adaptable to varying batch sizes without introducing redundant computations or excessive memory usage.

##### 14. Create patch embedding (equation 1)
```python
x = self.patch_embedding(x)
```
- This line converts the input image into a sequence of patch embeddings using the `PatchEmbedding` layer.
  - **Why transform the input image?** Transforming the image into embeddings enables efficient processing by the Transformer architecture, maintaining spatial relationships while reducing dimensionality. Embeddings serve as compact, informative representations suitable for subsequent layers that extract higher-level features [[3]].
  - **Process details:** Typically involves partitioning the image into non-overlapping patches, flattening them, and projecting onto the embedding space. This step bridges the gap between raw pixel data and abstract feature spaces, laying the groundwork for advanced modeling capabilities.

##### 15. Concat class embedding and patch embedding (equation 1)
```python
x = torch.cat((class_token, x), dim=1)
```
- This line concatenates the class embedding with the patch embeddings along the sequence dimension (`dim=1`).
  - **Why concatenate along `dim=1`?** Concatenating along the second dimension combines tokens sequentially, preserving their order for the Transformer to process effectively. This arrangement aligns with the expected input format for Transformer models, ensuring compatibility and seamless integration [[3]].
  - **Sequential nature:** Maintaining a clear sequence structure is vital for capturing dependencies and interactions among tokens, enabling the model to leverage contextual information comprehensively.

##### 16. Add position embedding to patch embedding (equation 1)
```python
x = self.position_embedding + x
```
- This line adds the position embeddings to the patch embeddings.
  - **Why add position embeddings?** Adding position embeddings informs the model about the relative positions of patches, crucial for capturing spatial hierarchies within the image. Unlike RNNs or CNNs, Transformers lack inherent notions of order, necessitating explicit positional encoding to preserve spatial context [[3]].
  - **Additive approach benefits:** Summing position and patch embeddings allows the model to jointly optimize both components, fostering synergistic learning and adaptive representation formation tailored to the specific task at hand.

##### 17. Run embedding dropout (Appendix B.1)
```python
x = self.embedding_dropout(x)
```
- This line applies dropout to the embeddings.
  - **Why apply dropout here?** Applying dropout after embedding introduces noise, preventing the model from relying too heavily on specific features and enhancing generalization. Targeting embeddings specifically mitigates risks associated with memorization or overfitting to particular tokens or positions, promoting robustness [[8]].
  - **Dropout impact:** Randomly zeroing elements induces variability in training samples, encouraging the model to develop resilient pathways that generalize better across diverse scenarios and inputs.

##### 18. Pass patch, position and class embedding through transformer encoder layers (equations 2 & 3)
```python
x = self.transformer_encoder(x)
```
- This line passes the embeddings through the Transformer encoder layers.
  - **Why multi-head attention and MLP?** Multi-head attention captures diverse relationships between tokens, while the MLP processes these representations further, enabling complex feature extraction. Combined, they facilitate comprehensive understanding and synthesis of input data, driving superior performance in various tasks [[3]].
  - **Layer progression:** Successive layers progressively refine representations, integrating local and global contexts to form rich, hierarchical features conducive to accurate downstream predictions.

##### 19. Put 0 index logit through classifier (equation 4)
```python
x = self.classifier(x[:, 0]) # run on each sample in a batch at 0 index
```
- This line passes the class token (which is at index 0 in the sequence) through the classifier to produce the final output logits.
  - **Why index 0?** Index 0 corresponds to the class token, which aggregates global information from all patches, serving as the summary representation for classification. Isolating this token ensures focused utilization of distilled knowledge encapsulated within it, maximizing discriminative power [[3]].
  - **Classification logic:** Mapping the aggregated representation to output classes leverages learned patterns and structures, translating latent insights into actionable predictions aligned with the target objective.

##### 20. Return the output
```python
return x
```
- Finally, the output logits are returned.
  - **Why return logits?** Logits represent the unnormalized predictions, which are typically passed to a loss function like cross-entropy for supervised learning tasks. Providing raw scores enables flexible handling and integration with various training regimes and evaluation metrics, supporting comprehensive assessment and optimization efforts [[8]].

2. Setup a ViT model instance via `torchvision.models.vit_b_16`, pass it the pretrained weights step 1 and send it to the target device.
3. Freeze all of the parameters in the base ViT model created in step 2 by setting their `requires_grad` attribute to `False`.
4. Update the classifier head of the ViT model created in step 2 to suit our own problem by changing the number of `out_features` to our number of classes (pizza, steak, sushi).

We covered steps like this in 06. PyTorch Transfer Learning [section 3.2: Setting up a pretrained model](https://www.learnpytorch.io/06_pytorch_transfer_learning/#32-setting-up-a-pretrained-model) and [section 3.4: Freezing the base model and changing the output layer to suit our needs](https://www.learnpytorch.io/06_pytorch_transfer_learning/#34-freezing-the-base-model-and-changing-the-output-layer-to-suit-our-needs).

````python
# 1. Get pretrained weights for ViT-Base
pretrained_vit_weights = torchvision.models.ViT_B_16_Weights.DEFAULT # requires torchvision >= 0.13, "DEFAULT" means best available

# 2. Setup a ViT model instance with pretrained weights
pretrained_vit = torchvision.models.vit_b_16(weights=pretrained_vit_weights).to(device)

# 3. Freeze the base parameters
for parameter in pretrained_vit.parameters():
    parameter.requires_grad = False

# 4. Change the classifier head (set the seeds to ensure same initialization with linear head)
set_seeds()
pretrained_vit.heads = nn.Linear(in_features=768, out_features=len(class_names)).to(device)
# pretrained_vit # uncomment for model output
````



- **Line 1: Get pretrained weights for ViT-Base**
  ```python
  pretrained_vit_weights = torchvision.models.ViT_B_16_Weights.DEFAULT
  ```
  - `torchvision.models.ViT_B_16_Weights.DEFAULT`: This line retrieves the default pretrained weights for the Vision Transformer (ViT) model with the `ViT-B/16` architecture. These weights are stored in the `torchvision` library, which is a PyTorch module for computer vision tasks.
    - **Why is `vit_b_16_weights` in `torchvision.models`?** The `torchvision.models` module contains predefined architectures and their pretrained weights for various models, including Vision Transformers (ViT). These weights are pre-trained on large datasets like ImageNet, making them useful for transfer learning.
    - **Are these weights already in the module called `torchvision`?** Yes, the weights are part of the `torchvision` library, and you can access them directly using the `ViT_B_16_Weights.DEFAULT` attribute.

- **Line 2: Setup a ViT model instance with pretrained weights**
  ```python
  pretrained_vit = torchvision.models.vit_b_16(weights=pretrained_vit_weights).to(device)
  ```
  - `torchvision.models.vit_b_16(weights=pretrained_vit_weights)`: This line creates an instance of the Vision Transformer model with the `ViT-B/16` architecture and loads the pretrained weights.
    - **Is `vit_b_16` a function?** Yes, `vit_b_16` is a function provided by `torchvision.models` that constructs a Vision Transformer model with the `ViT-B/16` architecture.
    - **Why is it a function?** The function is used to instantiate the model with specific configurations, such as the number of layers, hidden units, and other architectural details. It also allows you to load pretrained weights.
    - **What does this function do?** The `vit_b_16` function initializes the Vision Transformer model with the specified architecture and loads the pretrained weights if provided.
  - `.to(device)`: This moves the model to the specified device (e.g., CPU or GPU) for computation.

- **Line 3: Freeze the base parameters**
  ```python
  for parameter in pretrained_vit.parameters():
      parameter.requires_grad = False
  ```
  - **Why freeze the base parameters and set `requires_grad = False`?** Freezing the base parameters means that these parameters will not be updated during training. This is typically done in transfer learning scenarios where you want to retain the knowledge learned by the pretrained model on a large dataset (e.g., ImageNet) and only fine-tune the final layers for a specific task. Setting `requires_grad = False` ensures that the gradients are not computed for these parameters during backpropagation, which saves computation and memory.

- **Line 4: Change the classifier head**
  ```python
  set_seeds()
  pretrained_vit.heads = nn.Linear(in_features=768, out_features=len(class_names)).to(device)
  ```
  - `set_seeds()`: This function sets the random seed for reproducibility, ensuring that the initialization of the model's parameters is consistent across runs.
  - `pretrained_vit.heads = nn.Linear(in_features=768, out_features=len(class_names))`: This line replaces the classifier head of the Vision Transformer model with a new linear layer.
    - **What's the classifier head?** The classifier head is the final layer of the model that maps the learned features to the output classes. In this case, it is a linear layer that takes the output of the transformer and maps it to the number of classes in your dataset.
    - **Why set `in_features` to 768?** The `in_features=768` corresponds to the dimensionality of the output features from the Vision Transformer's encoder. For the `ViT-B/16` model, the output feature size is 768.
    - **Why set `out_features` to `len(class_names)`?** The `out_features` parameter is set to the number of classes in your dataset (`len(class_names)`). This ensures that the model outputs a prediction for each class in your specific task.
  - `.to(device)`: This moves the new classifier head to the specified device (e.g., CPU or GPU).

## Main takeaways

* With the explosion of machine learning, new research papers detailing advancements come out every day. And it's impossible to keep up with it *all* but you can narrow things down to your own use case, such as what we did here, replicating a computer vision paper for FoodVision Mini.
* Machine learning research papers often contain months of research by teams of smart people compressed into a few pages (so teasing out all the details and replicating the paper in full can be a bit of challenge).
* The goal of paper replicating is to turn machine learning research papers (text and math) into usable code.
    * With this being said, many machine learning research teams are starting to publish code with their papers and one of the best places to see this is at [Paperswithcode.com](https://paperswithcode.com/)
* Breaking a machine learning research paper into inputs and outputs (what goes in and out of each layer/block/model?) and layers (how does each layer manipulate the input?) and blocks (a collection of layers) and replicating each part step by step (like we've done in this notebook) can be very helpful for understanding.
* Pretrained models are available for many state of the art model architectures and with the power of transfer learning, these often perform *very* well with little data.
* Larger models generally perform better but have a larger footprint too (they take up more storage space and can take longer to perform inference).
    * A big question is: deployment wise, is the extra performance of a larger model worth it/aligned with the use case?
